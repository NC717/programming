# Doing Data Science

Data science is about finding the signals buried in the noise. Ultimately, the goal is to arrive at insights we can rely on to make high-quality decisions in the real world.

Practical methods of induction is the go-to method for this. It takes you from observations to hypotheses to the credibility of each hypothesis. In practice, you start with a hypothesis and collect data you think can give you answers. Then, you generate a model and use it to explain the data. Next, you evaluate the credibility of the model based on how well it explains the data observed so far.

## Step 1: Answering the question

The first step to any data analysis project is to define the question or problem we're looking to solve, and to define a measure (or set of measures) for our success at solving that task. Here are some useful checklists:

Specify the type of data analytic question before touching the data:

> We're trying to classify the species (i.e., class) of the flower based on four measurements that we're provided: sepal length, sepal width, petal length, and petal width.

Define the metric for success and the current baseline:

> Since we're performing classification, we can use accuracy — the fraction of correctly classified flowers — to quantify how well our model is performing. Our head of data has told us that we should achieve at least 90% accuracy. Note that if this is the first time conducting this project, the appropriate metric may change after exploring the data. Because maybe we've discovered that the data is imbalanced and that accuracy is no longer suitable for the evaluating success.
